```python
import matplotlib.pyplot as plt

#sample data for energy consumption over 6 months (in Mwh)
months=["Jan","Feb","Mar","Apr","May","Jun"]
energy_consumption = [1200, 1300, 1100, 1500, 1400, 1600]
#create a line plot
plt.plot(months, energy_consumption, marker='o')
#add labels and title
plt.xlabel("Months")
plt.ylabel("Energy Consumption (Mwh)")
plt.title("Energy Consumption Over 6 Months")
#show the plot
plt.show()

```

```python
#create a bar graph
plt.bar(months, energy_consumption)
#add labels and title
plt.xlabel("Months")
plt.ylabel("Energy Consumption (Mwh)")
plt.title("Energy Consumption Over 6 Months")
#show the plot
plt.show()
```

```python
# Assuming 'energy_values' represents the proportion of energy consumption for each source
# Replace these values with your actual data
energy_values = [12, 34, 30, 25]

# Assuming 'energy_sources' represents the names of the energy sources
# Replace these values with your actual data
energy_sources = ['Solar', 'Wind', 'Hydro', 'Fossil Fuels']

plt.pie(energy_values, labels=energy_sources, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen', 'coral'])
# add title
plt.title("Energy Consumption Share by source")
# show the plot
plt.show()
```

```python
#create scatter plot
plt.scatter(months, energy_consumption, color='r', marker='o')

#add titles and lables
plt.title('Energy Consumption Over 6 Months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (Mwh)')
```

```python
carbon_emission = [400, 500, 450, 300, 350, 550]

plt.scatter(energy_consumption, carbon_emission,color = 'red')

plt.title('Energy Consumption vs Carbon Emission')

plt.xlabel('Energy Consumption (Mwh)')

plt.ylabel('Carbon Emission (kg)')

plt.show()
```

```python
#pair plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

#creating a DataFrame for pair plot
data = {
    "Solar": [1200, 1500, 1300],
    "Wind": [3400, 3600, 3200],
    "Hydropower": [2900, 3100, 2800],
    "Biomass": [2500, 2700, 2400]
}

df=pd.DataFrame(data)

#creating a pair plot
sns.pairplot(df)

#showing the plot
plt.show()
```

```python
#heatmap : give correlational matrix
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',linewidth=0.5)
plt.title('Corelation between energy sources')
plt.show()
```

```python
#create a box plot
sns.boxplot(data=df)
plt.title('Box plot of energy sources')
plt.show()
```

```python
#create a violin plot
sns.violinplot(data=df)
plt.title('Violin plot of energy sources')
plt.show()
```

```python
#regression plot
#to understand relation betwn energy consumption and carbon emission
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

energy_values = [100,200,300,400,500]
carbon_emission = [10,20,25,40,50]

df_reg = pd.DataFrame({
    'energy_values': energy_values,
    'carbon_emission': carbon_emission
})

sns.regplot(x='energy_values', y='carbon_emission', data=df_reg)
plt.title('Relationship between energy consumption and carbon emission')
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure()
#syntax for 3-D projection
ax = plt.axes(projection='3d')

#defining all 3 axes
z = np.linspace(0, 1, 1000)
x= z * np.sin(25 * z)
y= z * np.cos(25 * z)

#plotiing

ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot')
plt.show()
```
